### Algoritmo de marcha aleatoria en 1d

```python
# Genera una marcha aleatoria en 1d, con el numero de iteraciones que entra por parametro
def genMarcha( n_steps ):
    # Posición del caminante en el eje único
    x = 0
    
    # Ciclo de los n pasos
    for i in range(n_steps):
        # Genera un numero aleatorio de 0 al 2 y decide hacia donde da el paso
        rand = 2*np.random.random()
        if ( (rand >= 0) & (rand < 1) ):
            x += 1
        else:
            x -= 1
        
    # Retorna la posición del caminante
    return x
```

### Generación de muestra estadística N = 8000

A proximación se generará una muestra 10000 marchas, cada una de 8000 pasos, y se comparará el histograma normalizado generado por esta muestra con la predicción del teorema central límite.

Este úlimo teorema, como se desarrolló en el punto $II$, dice que para $N$ grandes la probabilidad de que una partícula de $n$ pasos a la derecha se comporta de la siguiente manera:

$$P(n,N) \ \ \approx \ \ \sqrt[]{\frac{2}{N\pi}}  \ \ e^{-2\frac{(n-\frac{N}{2})^2}{N}}$$

Ahora,para encontrar esta misma probabilidad en función de $x$, la posición del caminante, se utiliza la relación demostrada en el punto $(2)$ como cambio de variable:

$$n\ \ = \ \ \frac{x}{2a} + \frac{N}{2} $$

y además se propone la relación de normalización:

$$ P(n,N) \ \ = \ \ kP(x,N) $$

Finalmente, tomando  $ a \  = \ 1 $ , y descubriendo que  $ k  \ = \  2 $ , se llega a la relación del teorema central límite que se puede comparar con los histogramas a graficar:

$$P(x,N) \ \ \approx \ \ \sqrt[]{\frac{1}{2N\pi}}  \ \ e^{-\frac{x^2}{2N}}$$

```python
# Devuelve la aproximación anterior del teorema central límite, dado un valor de x y deN
def tcl(x,N):
    return np.sqrt(1/(2*np.pi*N))*np.exp(-x*x/(2*N))
```

```python
# Guarda la distancia final de 10000 caminantes con caminatas de 8000 pasos
n_points = 10000
N = 8000
i = 0
R = np.zeros(n_points)
for i in range(n_points):
    R[i] = genMarcha(N)
```

```python
# Genera la gaussiana dada por tcl
xs = linspace(-400,400,500)
gauss = tcl(xs,N)
```

```python
# Grafica el histograma de la muestra y muestra la predicción del teorema central límite
histograma = hist(R, bins = 100,range = [-400,400], normed = True )
# Grafica la predicción por el teorema central límite
plt.title('Histograma y Probabilidad TCL')
plt.ylabel('Probabilidad')
plt.xlabel('Posicion')
plt.plot(xs,gauss,'r', linewidth = 2)
```

En el punto $III$ se demostró que le término |$r^2$| para un proceso de difusión en tres dimensiones está dado por $6Dt$.
Se puede demostrar por un proceso similar que para un proceso de difusión en una dimensión, como lo es la caminata que estamos tratando, el promedio de la posición al cuadrado está dado por:

|$r^2$| = $2Dt$
    
Si tomamos la convención del punto $II$, $ t = N\tau $ obtenemos:
    
|$r^2$| = $\frac{2DN}{\tau}$

Tomando $\tau = 1$, como se hizo con $a$, se obtiene:
    
|$r^2$| = $2DN$
    
En las siguientes lineas se correrá un número importante de caminatas variando $N$, para evidenciar las relaciones anteriormente encontradas.

```python
# Corre el programa de la caminata m veces para cada de N entre Nmin y Nmax, N a intervalos de dN (dN entero) y retorna <r^2> y <r> para cada grupo con N
def relat( Nmin, Nmax, dN, m ):
    
    # Inicializa las variables
    promR2 = []
    promR = []
    # El N actual
    Nactual = Nmin
    
    while( Nactual < Nmax ):
        i = 0
        R = 0
        R2 = 0
        # Genera m marchas
        for i in range(m): 
            a = genMarcha(Nactual)
            R += a
            R2 += a*a
            
        # Añade los valores promedio a la lista
        promR2.append(R2/m)
        promR.append(R/m)
        
        # Paso de iteración 
        Nactual += dN
        
    return [promR,promR2]
    
```

```python
# Corre el metodo relat para Nmin = 1000, Nmax = 10000, m = 100, dN = 40

values = relat(1000,10000,40,100)
```

```python
# Grafica los valores de <r> y <r^2>
Nspace = linspace(1000,10000,225)
plt.scatter(Nspace, np.array(values[0]), c = 'r', alpha = 0.5)
plt.scatter(Nspace, np.array(values[1]), c = 'b', alpha = 0.5)
plt.title(' <r> y <r^2> ')
plt.ylabel( '<r> (red), <r^2> (blue) ' )
plt.xlabel( 'N' )
           
```
